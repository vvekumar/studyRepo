# Signals - 01

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,1,10)  # a real sequence
plt.stem(x); 
plt.xlabel('Sample Index'); 
plt.ylabel('Amplitude');
plt.title('A Simple discrete time signal', fontweight='bold')
plt.grid()
plt.show()
```

**Sample Rate Alteration**

```python
'''Create a signal as F0 = 8kHz, and Fs = 16kHz'''
F0 = 8000
Fs = 16000
t_total = 1e-2 # sec
t = np.linspace(0,t_total, Fs*t_total)
x = np.sin(2*np.pi*F0*t)

# Get FFT Magnitude of original signal
fft_x = np.fft.fft(x)
mag_fft_x = abs(fft_x)

# Plot orginal
plt.subplot(211)
plt.stem(x)
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Sinusoid; F0 = 8kHz', fontweight='bold')

plt.subplot(212)
plt.plot(mag_fft_x)
plt.xlabel('FFT Bin')
plt.ylabel('Amplitude')
plt.show()
```

** Create new interpolated signal **

```python
# y = [val if idx % 8 == 0 else 99 for idx, val in enumerate(x) ]  # Preserves same length, but makes element zero
y = [val for idx, val in enumerate(x) if idx % 8 == 0]  # Shrinks length of the downsampled signal

# Get FFT Magnitude of original signal
fft_y = np.fft.fft(y)
mag_fft_y = abs(fft_y)

# Plot orginal
plt.subplot(211)
plt.stem(y)
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Sinusoid; F0 = 8kHz(factor of 8 downsampled)', fontweight='bold')

plt.subplot(212)
plt.plot(mag_fft_y)
plt.xlabel('FFT Bin')
plt.ylabel('Amplitude')
plt.show()
```
